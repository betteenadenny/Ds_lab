```python
# 1. Draw a line in a diagram from position (1, 3) to (2, 10) then to (6, 12) and finally to 
# position (18, 20). (Mark each point with a beautiful green colour and set line colour to red 
# and line style dotted) 
import numpy as np
import matplotlib.pyplot as plt
x = np.array([1,2,6,18])
y = np.array([3,10,12,20])
# plt.scatter(x,y)
plt.plot(x,y,c='r',linestyle='dotted',marker='o',mfc='g',mec='g')#mfc=marker face color mec=m edge c
plt.show()
```

```python
# 3. Write a Python program to draw a line using given axis values taken from a text file, 
# with suitable label in the x axis, y axis and a title. 
with open("data.txt",'r') as file:
    data = file.read().splitlines()
x_values = []
y_values = []
for line in data:
  x, y = map(float, line.split())     
  x_values.append(x)
  y_values.append(y)
plt.plot(x_values, y_values)
plt.show()
```

```python
# 6. Consider the following data.
# (i) Write a Python programming to display a bar chart of the popularity of programming 
# Languages. 
# (ii) Write a Python programming to display a horizontal bar chart of the popularity of 
# programming Languages(Give Red colour to the bar chart).
# (iii) Write a Python programming to display a bar chart of the popularity of programming 
# Languages. Use different color for each bar. 
import matplotlib.pyplot as plt

objects = ('Java','Python', 'PHP', 'javascript' , 'c#' , 'C++')
y_pos = np.arange(len(objects))
performance = [22.2,17.6,8.8,8,77,6.7]

plt.bar(y_pos, performance)
plt.xticks(y_pos, objects)
plt.ylabel('Popularity')
plt.title('Programming language usage')
plt.show()


plt.barh(y_pos, performance, align='center', alpha=0.5,color='r')
plt.yticks(y_pos, objects)
plt.xlabel('Popularity')
plt.title('Programming language usage')
plt.show()

bar=plt.bar(y_pos, performance,color=['green','blue','black','red','yellow','orange'])
plt.xticks(y_pos, objects)
plt.ylabel('Popularity')
plt.title('Programming language usage')
plt.show()
```

```python
# 7. Write a Python program to create bar plot of scores by group and gender. Use multiple 
# X values on the same chart for men and women. 
# Sample Data: 
# Means (men) = (22, 30, 35, 35, 26) and Means (women) = (25, 32, 30, 35, 29)

men_means = [22, 30, 35, 35, 26]
women_means = [25, 32, 30, 35, 29] 
index = np.arange(len(men_means))
bar_width = 0.35
plt.bar(index, men_means, bar_width,label='Men')
plt.bar(index + bar_width, women_means, bar_width,label='Women')
plt.xlabel('Person')
plt.ylabel('Scores')
plt.title('Scores by person')
plt.xticks(index + bar_width/2, ['G1', 'G2', 'G3', 'G4', 'G5'])
plt.legend()
plt.show()
```

```python
# 8. Write a Python programming to create a pie chart of the popularity of programming Languages.
languages = 'Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++'
popuratity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
plt.pie(popuratity,  labels=languages,autopct='%1.1f%%')
plt.show()
 
```

```python
# 9.Write a Python programming to create a pie chart of gold medal achievements of five 
# most successful countries in 2016 Summer Olympics. Read the data from a csv file. 
# Sample data: 
# medal.csv 
# country,gold_medal 
# United States,46 
# Great Britain,27 
# China,26 
# Russia,19 
# Germany,17
import pandas as pd
import matplotlib.pyplot as plt
# Read data from CSV file
df = pd.read_csv('data.csv')
countries = df['country']
gold_medals = df['medal']
plt.pie(gold_medals, labels=countries, autopct='%1.1f%%')
# Equal aspect ratio ensures that the pie is drawn as a circle.
plt.axis('equal')
plt.title('Gold Medal Achievements - 2016 Summer Olympics')
plt.show()
```
