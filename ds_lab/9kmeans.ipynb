```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(42)

# generates a synthetic dataset with 300 data points using the make_blobs function, and the resulting data points are stored in the variable data. The variable '_' is used as a placeholder for an element that we are not using in this context.
data, _ = make_blobs(n_samples=300)

model = KMeans(n_clusters=5, random_state=42, n_init="auto")# Automatically chooses the number of times the algorithm will be run with different centroid seeds.

labels = model.fit(data).labels_ # Retrieves the cluster labels assigned to each data point by the k-means algorithm. Each label corresponds to the cluster to which the corresponding data point belongs.
set(labels) # Creates a set containing the unique cluster labels. This is used to inspect the unique cluster labels assigned by k-means to the data points

centroids = model.fit(data).cluster_centers_

plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis', alpha=0.7, label='Cluster')
plt.scatter(centroids[:, 0], centroids[:, 1], color="red", marker="x", label='Centroids')
plt.scatter([2, 6], [-4, 6], color="red", label='New Points')
plt.legend()
plt.title('K-Means Clustering with New Points')
plt.show()
```

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt
import numpy as np
np.random.seed(42)

data,_ = make_blobs(n_samples=300)
model = KMeans(n_clusters=5,random_state=42,n_init='auto')

label = model.fit(data).labels_
set(label)
centroid = model.fit(data).cluster_centers_

plt.scatter(data[:,0],data[:,1],c=label,alpha=0.8,label='cluster')
plt.scatter(centroid[:,0],centroid[:,1],label ='centroid',marker='x',c='r')
plt.scatter([6, 3], [-4, 5], color="r", label='New Points')
plt.legend()
plt.show()
```
